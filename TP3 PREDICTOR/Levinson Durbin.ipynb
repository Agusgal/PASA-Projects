```
import numpy as np
def do_levinson_durbin(r):
  r = np.array(r)
  M = len(r)
  w = np.zeros_like(r)
  K = np.zeros_like(r)
  wn = np.zeros_like(w)
  alpha = 0
  #init 
  w[0] = 1
  K[0] = -r[1]/r[0]
  alpha = r[0]*(1-np.power(K[0],2))
  print(alpha)
  for i in range(2,M):
    S = np.sum([r[j]*w[i-j] for j in range(1,i)]) + r[i]
    K[i] = (-S)/alpha
    for j in range(1,i):
      wn[j] = w[j]+K[i]*w[i-j]
    wn[i] = K[i]+1e-5
    print(wn[i])
    alpha = alpha*(1-np.power(K[i],2))
    print(alpha)
  return w
```
