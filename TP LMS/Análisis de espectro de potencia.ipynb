```
import matplotlib.pyplot as plt
from matplotlib.mlab import psd, window_hanning
from scipy.io import wavfile
import numpy as np
```

```
from pathlib import PurePath
root = PurePath("/content/drive/Shared drives/PASA/TP LMS/Sonidos")
airplane_sound_path = root.joinpath("airplaneinterior.wav")
factory_fan_sound_path = root.joinpath("factoryfan.wav")
```

```
fs1, airplane_sound = wavfile.read(airplane_sound_path)
fs2, factory_fan_sound = wavfile.read(factory_fan_sound_path)
```

```
# Chequeamos si los sonidos son mono o stereo
print(airplane_sound.shape)
print(factory_fan_sound.shape)
```

## Avión

```
plt.plot(airplane_sound);
```

```
p_airplane, f_airplane = psd(x=airplane_sound, NFFT=2**12, Fs=fs1, detrend="mean")
plt.xlabel("freq $[Hz]$"); plt.ylabel("Power Density $[W/Hz]$"); plt.title("PSD airplane sound")
plt.semilogx(f_airplane, p_airplane);
```

## Fabrica

```
plt.plot(factory_fan_sound);
```

```
p_factory, f_factory = psd(x=factory_fan_sound, NFFT=2**12, Fs=fs1, detrend="mean")
plt.xlabel("freq $[Hz]$"); plt.ylabel("Power Density $[W/Hz]$"); plt.title("PSD factory fan sound")
plt.semilogx(f_factory ,p_factory);
```
